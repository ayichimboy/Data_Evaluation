```python
# import libraries
import numpy as anp            
import seaborn as sns                
import pandas as pd                   
import matplotlib.pyplot as plt                        
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,
                  n_features=2,
                  centers=4,
                  cluster_std=1.8,
                  random_state=42   
)

```

```python
# Kmeans Clustering 
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
# plot results
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], cmap='rainbow')
```

#### DBSCAN Clustering

```python
# read the csv data of the penguins
df = pd.read_csv('penguins.csv')
df.head()
```

```python
df.info()
```

```python
df_clean = df.dropna(subset='sex')
df_clean.info()
```

```python
sns.pairplot(data=df_clean, hue='species')
plt.show()
```

```python
df_clean.columns.to_list()
```

```python
from sklearn.cluster import DBSCAN

# define x and y variables
X = df_clean[['bill_length_mm', 'bill_depth_mm','flipper_length_mm']]
print(X.head())
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=4)
labels = dbscan.fit_predict(X)

df_clean['dbscan_labels'] = labels
sns.pairplot(data=df_clean, hue='dbscan_labels')
plt.show()

#print(labels)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df_clean['bill_length_mm'], df_clean['flipper_length_mm'], c=labels, cmap='plasma', s=60)
plt.xlabel('Bill Length (mm)')
plt.ylabel('Flipper Length (mm)')
plt.title('DBSCAN Clustering of Penguins')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
# kmeans clustering with penguins dataset
kmeans_penguins = KMeans(n_clusters=3)
kmodel = kmeans_penguins.fit(X)

df_clean['kmeans_labels'] = kmodel.labels_
#df_clean.head()
```

```python
# use PCA dimensionality reduction to plot the clusters
from sklearn.decomposition import PCA  

pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(X)

pca_transformed_df = pd.DataFrame(data=pca_transformed, columns=['pca1', 'pca2'])
df_clean = pd.concat([df_clean, pca_transformed_df], axis=1)

df_clean.head()
# print(pca_transformed)
# df_clean['pca']
```

```python
sns.scatterplot(data=df_clean, x='bill_length_mm', y='flipper_length_mm', hue='kmeans_labels', palette='plasma')
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize = (10,6))

ax1.set_title('Original Cluster')
sns.scatterplot(data=df_clean, x='bill_length_mm', y='flipper_length_mm', hue='species', palette='plasma', ax=ax1)

ax2.set_title('Kmeans Cluster')
sns.scatterplot(data=df_clean, x='bill_length_mm', y='flipper_length_mm', hue='kmeans_labels', palette='plasma', ax=ax2)

plt.show()

# ax3.set_title('PCA Cluster')
# sns.scatterplot(data=df_clean, x='pca1', y='pca2', hue='species', palette='plasma', ax=ax3)
```

```python
# plot PCA clusters
plt.figure(figsize=(8, 6))
plt.scatter(df_clean['pca1'], df_clean['pca2'], c=df_clean['kmeans_labels'], cmap='plasma', s=60)
plt.show()
```
